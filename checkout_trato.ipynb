```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_excel(r'carrinho_abandonado_.xlsx')
display(df)
```

```python
display(df[['checkOut']])
```

```python
df_checkout = df['checkOut'].str.split(',', expand=True)
df_checkout
```

```python
df_checkout.info()
```

```python
def lmp_char(df):
    df=df.replace('{','',regex=True)
    df=df.replace('}','',regex=True)
    df=df.replace("'",'',regex=True)
    
    return df
```

```python
df_checkout = lmp_char(df_checkout)
df_checkout
```

```python
def to_str(df):
    df.columns = df.columns.map(str)
    return df
```

```python
df_checkout = to_str(df_checkout)
df_checkout
```

```python
df_checkout['0'] = df_checkout['0'].str.replace('_id: ', '')
df_checkout['1'] = df_checkout['1'].str.replace('date: ', '')
df_checkout['2'] = df_checkout['2'].str.replace('time: ', '')


df_checkout = df_checkout.rename(columns={'0': 'id_checkout', '1': 'data_checkout', '2': 'hora_checkout'})

display(df_checkout)
```

```python
def l_space(df):
    df = df.str.strip()
    return df
```

```python
df_checkout['hora_checkout'] = l_space(df_checkout['hora_checkout'])
```

```python
df_checkout.to_excel('tabela_checkout.xlsx')
```
