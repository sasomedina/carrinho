```python
def expand(df):
    df = df.str.split(',', expand=True)
    return df
```

```python
df_status = expand(df['statuses'])
df_status
```

```python
df_status.info(max_cols=105)
```

```python
df_status.columns = df_status.columns.map(str)
df_status
```

```python
df_status = df_status[['0', '1', '2', '3', '4', '5']]
df_status
```

```python
def lmp_char(df):
    df=df.replace('{','',regex=True)
    df=df.replace('}','',regex=True)
    df=df.replace('\[','',regex=True)
    df=df.replace(']','',regex=True)
    df=df.replace("'",'',regex=True)
    return df
```

```python
df_status = lmp_char(df_status)
df_status
```

```python
df_status['0'] = df_status['0'].str.replace('name: ', '')
df_status['1'] = df_status['1'].str.replace('createdAt: ', '')
df_status['2'] = df_status['2'].str.replace('_id: ', '')
df_status['3'] = df_status['3'].str.replace('name: ', '')
df_status['4'] = df_status['4'].str.replace('createdAt: ', '')
df_status['5'] = df_status['5'].str.replace('_id: ', '')


df_status = df_status.rename(columns={'0': 'status_1', '1': 'data_status_1', '2': 'status_1_id', '3': 'status_2', 
                                      '4': 'data_status_2', '5': 'status_2_id'})

display(df_status)
```

```python
def l_space(df):
    df = df.str.strip()
    return df
```

```python
df_status['status_1'] = l_space(df_status['status_1'])
df_status['data_status_1'] = l_space(df_status['data_status_1'])
df_status['status_1_id'] = l_space(df_status['status_1_id'])
df_status['status_2'] = l_space(df_status['status_2'])
df_status['data_status_2'] = l_space(df_status['data_status_2'])
df_status['status_2_id'] = l_space(df_status['status_2_id'])
```

```python
df_status.to_excel('tabela_status.xlsx')
```
