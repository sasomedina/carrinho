```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df_user = df['user'].str.split(',', expand=True)
df_user
```

```python
df_user.info()
```

```python
def lmp_char(df):
    df=df.replace('{','',regex=True)
    df=df.replace('}','',regex=True)
    df=df.replace('\[','',regex=True)
    df=df.replace(']','',regex=True)
    df=df.replace("'",'',regex=True)
    return df
```

```python
df_user = lmp_char(df_user)
df_user
```

```python
df_user.columns = df_user.columns.map(str)
df_user
```

```python
df_user['0'] = df_user['0'].str.replace('_id: ', '')
df_user['1'] = df_user['1'].str.replace('staysId: ', '')
df_user['2'] = df_user['2'].str.replace('firstName: ', '')
df_user['3'] = df_user['3'].str.replace('lastName: ', '')
df_user['4'] = df_user['4'].str.replace('email: ', '')


df_user = df_user.rename(columns={'0': 'user_id', '1': 'stay_id', '2': 'nome_user', '3': 'sobrenome_user', '4': 'email'})

display(df_user)
```

```python
def l_space(df):
    df = df.str.strip()
    return df
```

```python
df_user['user_id'] = l_space(df_user['user_id'])
df_user['stay_id'] = l_space(df_user['stay_id'])
df_user['nome_user'] = l_space(df_user['nome_user'])
df_user['sobrenome_user'] = l_space(df_user['sobrenome_user'])
df_user['email'] = l_space(df_user['email'])
```

```python
df_user.to_excel('tabela_user.xlsx')
```
