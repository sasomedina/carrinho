```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_excel(r'carrinho_abandonado_.xlsx')
display(df)
```

```python
df.info()
```

```python
display(df[['partner']])
```

```python
df_partner = df['partner'].str.split(',', expand=True)
df_partner
```

```python
df_partner.info()
```

```python
df_partner=df_partner.replace('{','',regex=True)
df_partner=df_partner.replace('}','',regex=True)
df_partner=df_partner.replace('\[','',regex=True)
df_partner=df_partner.replace(']','',regex=True)
df_partner=df_partner.replace("'",'',regex=True)
df_partner
```

```python
df_partner.columns = df_partner.columns.map(str)
```

```python
df_partner['0'] = df_partner['0'].str.replace('integrationType: ', '')
df_partner['1'] = df_partner['1'].str.replace('_id: ', '')
df_partner['2'] = df_partner['2'].str.replace('aliasId: ', '')
df_partner['3'] = df_partner['3'].str.replace('name1: ', '')
df_partner['4'] = df_partner['4'].str.replace('name2: ', '')

df_partner = df_partner.rename(columns={'0': 'tipo_integração', '1': 'id_partner', '2': 'alias_partner', '3': 'nome_gestor', 
                                         '4': 'nome_partner'})

display(df_partner)
```

```python
def l_space(df):
    df = df.str.strip()
    return df
```

```python
df_partner['tipo_integração'] = l_space(df_partner['tipo_integração'])
df_partner['id_partner'] = l_space(df_partner['id_partner'])
df_partner['alias_partner'] = l_space(df_partner['alias_partner'])
df_partner['nome_gestor'] = l_space(df_partner['nome_gestor'])
df_partner['nome_partner'] = l_space(df_partner['nome_partner'])
```

```python
df_partner.to_excel('tabela_partner.xlsx')
```
