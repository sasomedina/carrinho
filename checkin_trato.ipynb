```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
display(df[['checkIn']])
```

```python
df_checkin = df['checkIn'].str.split(',', expand=True)
df_checkin
```

```python
df_checkin.info()
```

```python
def lmp_char(df):
    df=df.replace('{','',regex=True)
    df=df.replace('}','',regex=True)
    df=df.replace("'",'',regex=True)
    
    return df
```

```python
df_checkin = lmp_char(df_checkin)
df_checkin
```

```python
df_checkin
```

```python
def to_str(df):
    df.columns = df.columns.map(str)
    return df
```

```python
df_checkin = to_str(df_checkin)
df_checkin
```

```python
df_checkin['0'] = df_checkin['0'].str.replace('_id: ', '')
df_checkin['1'] = df_checkin['1'].str.replace('date: ', '')
df_checkin['2'] = df_checkin['2'].str.replace('time: ', '')


df_checkin = df_checkin.rename(columns={'0': 'id_checkin', '1': 'data_checkin', '2': 'hora_checkin'})

display(df_checkin)
```

```python
df_checkin.info()
```

```python
def l_space(df):
    df = df.str.strip()
    return df
```

```python
df_checkin['hora_checkin'] = l_space(df_checkin['hora_checkin'])
```

```python
df_checkin.to_excel('tabela_checkin.xlsx')
```
