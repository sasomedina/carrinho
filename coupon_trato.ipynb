```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_excel(r'carrinho_abandonado_.xlsx')
display(df.head())
```

```python
display(df[['coupon']])
```

```python
df[['coupon']].isna().sum()
```

```python
def expand(df):
    df = df.str.split(',', expand=True)
    return df
```

```python
df_coupon = expand(df['coupon'])
df_coupon
```

```python
def lmp_char(df):
    df=df.replace('{','',regex=True)
    df=df.replace('}','',regex=True)
    df=df.replace('\[','',regex=True)
    df=df.replace(']','',regex=True)
    df=df.replace("'",'',regex=True)
    return df
```

```python
df_coupon = lmp_char(df_coupon)
df_coupon
```

```python
df_coupon.columns = df_coupon.columns.map(str)
df_coupon.head()
```

```python
df_coupon['0'] = df_coupon['0'].str.replace('_id: ', '')
df_coupon['1'] = df_coupon['1'].str.replace('kind: ', '')
df_coupon['2'] = df_coupon['2'].str.replace('amount: ', '')
df_coupon['3'] = df_coupon['3'].str.replace('discountValue: amount: ', '')
df_coupon['4'] = df_coupon['4'].str.replace('currency: ', '')
df_coupon['5'] = df_coupon['5'].str.replace('precision: ', '')
df_coupon['6'] = df_coupon['6'].str.replace('originalValue: amount: ', '')
df_coupon['7'] = df_coupon['7'].str.replace('currency: ', '')
df_coupon['8'] = df_coupon['8'].str.replace('precision: ', '')
df_coupon['9'] = df_coupon['9'].str.replace('finalValue: amount: ', '')
df_coupon['10'] = df_coupon['10'].str.replace('currency: ', '')
df_coupon['11'] = df_coupon['11'].str.replace('precision: ', '')
df_coupon['12'] = df_coupon['12'].str.replace('name: ', '')


df_coupon = df_coupon.rename(columns={'0': 'cupom_id', '1': 'tipo', '2': 'amount', '3': 'valor_desconto', '4': 'moeda_desconto', 
                                     '5': 'precision_desconto', '6': 'valor_original', '7': 'moeda_valor_original', '8': 'precision_original', 
                                     '9': 'valor_final', '10': 'moeda_valor_final', '11': 'precision_final', '12': 'nome_cupom'})

display(df_coupon.head())
```

```python
def l_space(df):
    df = df.str.strip()
    return df
```

```python
df_coupon['cupom_id'] = l_space(df_coupon['cupom_id'])
df_coupon['tipo'] = l_space(df_coupon['tipo'])
df_coupon['amount'] = l_space(df_coupon['amount'])
df_coupon['valor_desconto'] = l_space(df_coupon['valor_desconto'])
df_coupon['moeda_desconto'] = l_space(df_coupon['moeda_desconto'])
df_coupon['precision_desconto'] = l_space(df_coupon['precision_desconto'])
df_coupon['valor_original'] = l_space(df_coupon['valor_original'])
df_coupon['moeda_valor_original'] = l_space(df_coupon['moeda_valor_original'])
df_coupon['precision_original'] = l_space(df_coupon['precision_original'])
df_coupon['valor_final'] = l_space(df_coupon['valor_final'])
df_coupon['moeda_valor_final'] = l_space(df_coupon['moeda_valor_final'])
df_coupon['precision_final'] = l_space(df_coupon['precision_final'])
df_coupon['nome_cupom'] = l_space(df_coupon['nome_cupom'])
```

```python
df_coupon.to_excel('tabela_cupom.xlsx')
```
