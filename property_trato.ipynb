```python
df_property['id_property'] = df_property['id_property'].str.replace('_id: ', '')
df_property['id_alias'] = df_property['id_alias'].str.replace('aliasId: ', '')
df_property['external_id'] = df_property['external_id'].str.replace('externalId: ', '')
df_property['nome_property'] = df_property['nome_property'].str.replace('name: ', '')
df_property['alias_property'] = df_property['alias_property'].str.replace('aliasName: ', '')
df_property['main_photo'] = df_property['main_photo'].str.replace('mainPhoto: ', '')
df_property['cidade'] = df_property['cidade'].str.replace('name: ', '')
df_property['id_bairro'] = df_property['id_bairro'].str.replace('neighborhood: _id: ', '')
df_property['bairro'] = df_property['bairro'].str.replace('name: ', '')
df_property['latitude'] = df_property['latitude'].str.replace('coordinates: ', '')
df_property['id_pais'] = df_property['id_pais'].str.replace('_id: ', '')
df_property['pais'] = df_property['pais'].str.replace('country: ', '')
df_property['estado'] = df_property['estado'].str.replace('state: ', '')
df_property['endereço'] = df_property['endereço'].str.replace('street: ', '')
df_property['número'] = df_property['número'].str.replace('streetNumber: ', '')
df_property['CEP'] = df_property['CEP'].str.replace('zipCode: ', '')
df_property = df_property.drop(columns=['6'])
```
