```python
dfcheckin = pd.read_excel(r'tabela_checkin.xlsx')
dfcheckout = pd.read_excel(r'tabela_checkout.xlsx')
display(dfcheckin)
display(dfcheckout)
```

```python
dfguests = pd.read_excel(r'tabela_guests.xlsx')
dfuser = pd.read_excel(r'tabela_user.xlsx')
dfcost = pd.read_excel(r'tabela_cost.xlsx')
display(dfguests)
display(dfuser)
display(dfcost)
```
