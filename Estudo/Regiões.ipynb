```python
# Criando um dataframe apenas com linhas booked

df_bkd = df[df['status'] == "BOOKED"]
df_bkd = df_bkd.drop(columns=['Unnamed: 0.1', 'Unnamed: 0.1.1', 'Unnamed: 0.1.1.1'])
df_bkd
```

```python
# Criando um dataframe apenas com linhas booked_manually

df_mbkd = df[df['status'] == "BOOKED_MANUALLY"]
df_mbkd = df_mbkd.drop(columns=['Unnamed: 0.1', 'Unnamed: 0.1.1', 'Unnamed: 0.1.1.1'])
df_mbkd
```

```python
# Juntando as duas tabelas a fim de ter ambos os tipos de booked

bkd = pd.concat([df_bkd, df_mbkd], axis=0)
bkd
```

```python
# Bairros com o maior número de "booked"

gb_bkd = bkd.groupby('bairro', as_index=True)['status'].count().sort_values(ascending=False)
gb_bkd
```

```python
gb_bkd.shape
```

```python
ax = gb_bkd.plot.bar(figsize=(19,8), rot=45)
ax.set_xlabel('Bairros', fontdict={'fontsize':18})
ax.set_ylabel('Total booked', fontdict={'fontsize':18})
ax.set_title('Total booked por bairro', fontdict={'fontsize':18})
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
fig.tight_layout()
```

```python
# Bairros com maior número de booked e suas respectivas vendas totais e médias.

pd.options.display.max_rows = 40
bkd_ = bkd.groupby('bairro', as_index=False).agg(
   total_vendas=pd.NamedAgg(column='cost_total', aggfunc=sum),
   media_vendas=pd.NamedAgg(column='cost_total', aggfunc=np.mean),
   status=pd.NamedAgg(column='status', aggfunc='count')).sort_values(by='status', ascending=False)
bkd_
```

```python
ax = bkd_.plot.bar(x='bairro', y='total_vendas', figsize=(19,8), rot=45)
ax.set_xlabel('Bairros', fontdict={'fontsize':18})
ax.set_ylabel('Total vendas', fontdict={'fontsize':18})
ax.set_title('Bairros com maior número de booked e suas respectivas vendas totais', fontdict={'fontsize':18})
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
fig.tight_layout()
```

```python
data_try = pd.get_dummies(data, columns=['status'])
data_try
```

```python
df_mail = data_try.groupby('guest_email', as_index=False).agg({'status_EXPIRED':'sum', 'status_BOOKED': 'sum', 'status_BOOKED_MANUALLY': 'sum'}).sort_values(by='status_EXPIRED', ascending=False)
df_mail
```

```python
# Tentativas por email. O número de tentativas por booked

df_mail['Tentativas'] = df_mail['status_EXPIRED'] / (df_mail['status_BOOKED'] + df_mail['status_BOOKED_MANUALLY'])
df_mail
```

```python
df_tentativas = df_mail.dropna(subset=['Tentativas'], how='any')
df_tentativas.sort_values(by='Tentativas', ascending=False)
```

```python
df_tentativas.replace([np.inf, -np.inf], np.nan, inplace=True)
df_tentativas
```

```python
df_tentativas.dropna(subset=['Tentativas'], how="any")
df_tentativas
```

```python
df_tentativas.sort_values(by='Tentativas', ascending=False)
```

```python
# Número de tentativas que cada email precisou para "bookar" no site.


df_tentativas['Tentativas'] = df_tentativas['Tentativas'].dropna()

df_tentativas.sort_values(by='Tentativas', ascending=False)
```

```python
# Tentativas por bairro. O número de tentativas por booked

df_reg['Tentativas'] = df_reg['status_EXPIRED'] / (df_reg['status_BOOKED'] + df_reg['status_BOOKED_MANUALLY'])
df_reg
```
